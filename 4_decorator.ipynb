```python
%%plantuml
@startuml

abstract class Beverage{
    -description = "unknown beverage": String
    +getDescription():Stirng
    +abstract cost():double
    }
class DarkRoast{
    +DarkRoast()
    +cost(): double
    }
class HouseBlend{
    +HouseBlend()
    +cost(): double
    }
abstract class CondimentDecorator{
    #beverage:Beverage
    +abstract getDescription(): String
    }
class Mocha{
    +Mocha()
    +getDescription():String
    +cost():double
    }
class Whip{
    +Whip()
    +getDescription():String
    +cost():double
    }
    
Beverage <|-- DarkRoast
Beverage <|-- HouseBlend
Beverage <|--CondimentDecorator
CondimentDecorator <|-- Mocha
CondimentDecorator <|-- Whip
Beverage <--* CondimentDecorator

@enduml
```

```python
%%plantuml
@startuml
group decorator DarkRoast
Client -> Beverage:new()
Client -> DarkRoast:beverage2=new()
Client -> Mocha:beverage2 = new(beverage2)
Client -> Mocha:beverage2 = new(beverage2)
Client -> Whip:beverage2 = new(beverage2)
Client -> Beverage :getCost()
end group
@enduml
```

```python
%%plantuml
@startuml

interface Window{
    +draw():void
    +getDescription():Stirng    
}

class SimpleWindow{
    +draw():void
    +getDescription():Stirng
}

abstract class WindowDecorator{
    # WindowDecorated:Window
    +WindowDecorator()
}

class VerticalScrollBarDecorator{
    +VerticalScrollBarDecorator()
    +draw():void
    +drawVerticalScrollBar()
    +getDescription():Stirng
}
class HorizontalScrollBarDecorator{
    +HorizontalScrollBarDecorator()
    +draw():void
    +drawHorizontalScrollBar()
    +getDescription():String
}

Window <|..SimpleWindow
Window <|..WindowDecorator
Window <--* WindowDecorator
WindowDecorator <|-- VerticalScrollBarDecorator 
WindowDecorator <|-- HorizontalScrollBarDecorator 


@enduml
```
