```python
%%plantuml

@startuml

class Client
class AutomaticDoor {
    +pressOpen()
}
class SlidingDoor {
    +slideOpen()
}
class KnobDoor {
    +twistOpen()
}
Client -up-> AutomaticDoor:pressOpen()
Client --> SlidingDoor:slideOpen()
Client -right-> KnobDoor:twistOpen()
    
@enduml
```

```python
%%plantuml

@startuml

class Client
class Door{
    +open()
}

class AutomaticDoor
class SlidingDoor 
class KnobDoor

Door <|-- AutomaticDoor
Door <|-- SlidingDoor
Door <|-- KnobDoor
Client -right->Door

@enduml
```

```python
%%plantuml

@startuml

Client -- Door
Door <|-- AutomaticDoor
Door <|-- KnobDoor
Door <|-- SlidingDoor

class Door {
  -isOpen
  +open()
}

note right of Client
    Client uses(depends) an abstract class 'Door'
    Previously, it depends on concrete classes.
end note
@enduml
```
